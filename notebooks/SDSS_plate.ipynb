#  Compact all files from same plate

```python
def addExtension(data, extname):
    from astropy.io import fits
    hdu = fits.ImageHDU()
    hdu.data = data
    hdu.header['EXTNAME']=(extname)
    return hdu
```

```python
from glob import glob as gb
from astropy.io import fits

plate = 655
path = '/media/Data/workspace/Clusters/A85/Spectra/SDSS/Florence/Data/'
files = gb(path+'spec-0'+str(plate)+'-*.fits')

minlw = []
maxlw = []
for file in files:
    with fits.open(file) as hdl:
        header = hdl['PRIMARY'].header
        data = hdl['COADD'].data
        lw = data['loglam']
        minlw.append(np.nanmin(lw))
        maxlw.append(np.nanmax(lw))
```

```python
nfiles = len(files)
dw = 0.0001
lwstart = np.nanmin(minlw)
lwend   = np.nanmax(maxlw)
lwgrid = np.arange(lwstart, lwend+dw, dw)
lwgrid = np.around(lwgrid, 4)
nw = len(lwgrid)
sky = np.zeros(nw)
err = np.zeros((nw,nfiles))
flux = np.zeros((nw,nfiles))
```

```python
fiber_id = []
fiber_ra = []
fiber_dec = []

for i,file in enumerate(files):
    with fits.open(file) as hdl:
        header = hdl['PRIMARY'].header
        data = hdl['COADD'].data
        flux_ = data['flux']
        lw_ = data['loglam']
        nw_ = len(lw_)
        sky_ = data['sky']
        err_ = np.sqrt(1./data['ivar'])
        fiber_id.append(header['FIBERID'])
        fiber_ra.append(header['PLUG_RA'])
        fiber_dec.append(header['PLUG_DEC'])
        start = int((np.min(lw_) - np.min(lwgrid))/dw)
        err[start:start+nw_, i] = err_
        flux[start:start+nw_, i] = flux_
        sky[start:start+nw_] = sky_
        
```

```python
hdu = fits.PrimaryHDU()
hdu.header["PLATEID"] = plate
for i, (fid, fra, fdec) in enumerate(zip(fiber_id, fiber_ra, fiber_dec)):
    hdu.header["FIBER_{:03d}_ID".format(i)] = fid 
    hdu.header["FIBER_{:03d}_RA".format(i)] = fra
    hdu.header["FIBER_{:03d}_DEC".format(i)] = fdec 
    
hdu1 = addExtension(flux, 'flux')
hdu2 = addExtension(err, 'err')
hdu3 = addExtension(sky, 'sky')
hdu4 = addExtension(lwgrid, 'logwave')
hdul = fits.HDUList([hdu, hdu1, hdu2, hdu3, hdu4])
hdul.writeto('/media/Data/workspace/Clusters/A85/Spectra/SDSS/SDSS_PLATE'+str(plate)+'.fits', overwrite=True) 
hdul.close()
```
